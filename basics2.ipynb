## Processing data

```python
questions = [str(x) for x in range(12, 20)]
```

```python
criteria = ['SUCURSAL', 'DEPARTAMENTO', 'SEXO', 'RANGO_EDAD', 'NIVEL_ESCOLAR_FINALIZADO']
```

```python
for q in questions:
    q_str = col_names_dict[int(q)]
    print('Q' + q + ': ' + q_str)
    print('*' * 80)
    print()
    q_df = survey_df.drop_duplicates(['NOMBRE', q])
    for c in criteria:
        print('By ' + c + ' - ' + q_str)
        print('-' * 80)
        display(pd.DataFrame(q_df.groupby(c)[q].value_counts()))
        print()
    print()
```
